```python
import sys, os
if os.path.abspath("..") not in sys.path:
    sys.path.insert(0, os.path.abspath(".."))

import matplotlib.pyplot as plt

from src.utils import find_popular_articles
from src.loader import NewsDataLoader

loader = NewsDataLoader()
popular_countries_data = loader.load_data('../data/data.csv/rating.csv')

popular_articles = find_popular_articles(popular_countries_data)
print(popular_articles)

countries = [country for (country, _), count in popular_articles]
counts = [count for (_, _), count in popular_articles]

plt.figure(figsize=(10, 6))
plt.bar(countries, counts, color='red')
plt.title('Top 10 Countries Mentioned in Articles')
plt.xlabel('Country')
plt.ylabel('Count')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

the following plot will show the sentiment categories of the news articles categorized by domain

```python
import sys, os
if os.path.abspath("..") not in sys.path:
    sys.path.insert(0, os.path.abspath(".."))

import matplotlib.pyplot as plt

from src.utils import website_sentiment_distribution
from src.loader import NewsDataLoader

loader = NewsDataLoader()
sentiment_data = loader.load_data('../data/data.csv/rating.csv')

sentiment_counts = website_sentiment_distribution(sentiment_data)

top_domains = sentiment_counts.sort_values(by='Total',ascending=False).head(10)

# Plot the mean sentiment counts of the top domains
top_domains[['Mean']].plot(kind='bar',figsize=(10,6))
plt.title('Mean Sentiment Counts of Top 10 Domains')
plt.xlabel('Domain')
plt.ylabel('Mean Sentiment Count')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()


# Plot the median sentiment counts of the top domains
top_domains[['Median']].plot(kind='bar', figsize=(10, 6))
plt.title('Median Sentiment Counts of Top 10 Domains')
plt.xlabel('Domain')
plt.ylabel('Median Sentiment Count')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
import sys, os
if os.path.abspath("..") not in sys.path:
    sys.path.insert(0, os.path.abspath(".."))

import matplotlib.pyplot as plt

from src.utils import webiste_sentiment
from src.loader import NewsDataLoader

loader = NewsDataLoader()
sentiment_data = loader.load_data('../data/data.csv/rating.csv')

sentiment_counts = webiste_sentiment(sentiment_data)

sentiment_counts.plot(kind='bar',stacked=True,figsize=(12,8))

plt.title('Sentiment Distribution by website domain')
plt.xlabel('Website Domain')
plt.ylabel('Number of Articles')

plt.xticks(rotation=45,ha='right')
plt.legend(title='Sentiment')

plt.tight_layout()
plt.show()
```
