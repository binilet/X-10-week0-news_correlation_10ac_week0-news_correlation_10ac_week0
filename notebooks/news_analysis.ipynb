```python
import sys, os
if os.path.abspath("..") not in sys.path:
    sys.path.insert(0, os.path.abspath(".."))

import matplotlib.pyplot as plt

from src.loader import NewsDataLoader
from bertopic import BERTopic
import pandas as pd

from nltk.corpus import stopwords
import nltk
nltk.download('stopwords')

seed_topics = {
    0: "Breaking News",
    1: "Politics",
    2: "World News",
    3: "Business/Finance",
    4: "Technology",
    5: "Science",
    6: "Health",
    7: "Entertainment",
    8: "Sports",
    9: "Environment",
    10: "Crime",
    11: "Education",
    12: "Weather",
    13: "Other"
}



loader = NewsDataLoader()
news_data = pd.read_csv('../data/data.csv/rating.csv')
sampled_data = news_data.sample(1000,random_state=42)
sampled_data['text'] = sampled_data['title'] + ' ' + sampled_data['content']


sampled_list = sampled_data['text'].tolist()

stop_words = set(stopwords.words('english'))

def remove_stopwords(text):
    words = text.split()
    filtered_words = [word for word in words if word.lower() not in stop_words]
    return ' '.join(filtered_words)

sampled_list = [remove_stopwords(text) for text in sampled_list]

topic_model = BERTopic()
topics,probs = topic_model.fit_transform(sampled_list)
print('done fitting ...')
topic_model.get_topic_info()
```
